```python
dat=data_v["yorumlar"].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
dat=pd.DataFrame(dat,columns=["yorumlar"])
```

```python
sat=dat["yorumlar"].apply(lambda x: " ".join(x for x in x.split() if x not in swt))
```

```python
sat
```

## Yorumları  durumuna göre kategorilere ayırma

```python
sat=pd.DataFrame(sat,columns=['yorumlar'])
```

```python
sat['positivity']=1
```

```python
sat['positivity'].replace(1,value = "pozitif", inplace = True)
```

```python
sat.positivity.iloc[146:] = 0
```

```python
sat['positivity'].replace(0,value = "negatif", inplace = True)
```

```python
sat
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(sat['yorumlar'], sat['positivity'], test_size = 0.2 ,random_state = 0)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(encoding ='iso-8859-9').fit(X_train)
```

```python
X_train_vectorized = vect.transform(X_train) 
```

```python
#Bu özellik matrisi X_ train_ vectorized'e dayanarak Lojistik Regresyon sınıflandırıcısını eğiteceğiz

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score
predictions = model.predict(vect.transform(X_test))
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names())
sorted_coef_index = model.coef_[0].argsort()
print('Negatif: \n{}\n'.format(feature_names[sorted_coef_index[:20]]))
print('Pozitif: \n{}\n'.format(feature_names[sorted_coef_index[:-20:-1]]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(min_df = 5).fit(X_train)
```

```python
X_train_vectorized = vect.transform(X_train)
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
predictions = model.predict(vect.transform(X_test))
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python

feature_names = np.array(vect.get_feature_names())
sorted_tfidf_index = X_train_vectorized.max(0).toarray()[0].argsort()
print('En küçük Tfidf: \n{}\n'.format(feature_names[sorted_tfidf_index[:20]]))
print('En büyük Tfidf: \n{}\n'.format(feature_names[sorted_tfidf_index[:-20:-1]]))
```

```python
vect = CountVectorizer(min_df = 5, ngram_range = (1,2)).fit(X_train)
X_train_vectorized = vect.transform(X_train)
```

```python
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
predictions = model.predict(vect.transform(X_test))
print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
#Her özelliği kontrol etmek için katsayıları kullanarak görebiliriz
feature_names = np.array(vect.get_feature_names())
sorted_coef_index = model.coef_[0].argsort()
print('Negatif: \n{}\n'.format(feature_names[sorted_coef_index][:20]))
print('Pozitif Coef: \n{}\n'.format(feature_names[sorted_coef_index][:-20:-1]))
```
